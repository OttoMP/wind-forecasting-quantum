```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(14,5), dpi=320, facecolor='w', edgecolor='k')
plt.title(f"Loss by depth")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid()
for n_layers in range(1,9):
    history = pd.read_csv(f"analysis/loss-df-{n_layers}.csv")
    plt.plot(history['Loss'], label=f"Loss depth {n_layers}")
    plt.xticks(range(0, len(history['Loss'])+1, 5))
    plt.legend(loc='best')
    #plt.plot(history['Val Loss'], label="Val Loss/Epoch")
```

```python
plt.figure(figsize=(14,5), dpi=320, facecolor='w', edgecolor='k')
plt.title(f"Val Loss by depth")
plt.xlabel("Epoch")
plt.ylabel("Val Loss")
plt.grid()
for n_layers in range(1,9):
    history = pd.read_csv(f"analysis/loss-df-{n_layers}.csv")
    plt.plot(history['Val Loss'], label=f"Val Loss depth {n_layers}")
    plt.xticks(range(0, len(history['Val Loss'])+1, 5))
    plt.legend(loc='best')
```
