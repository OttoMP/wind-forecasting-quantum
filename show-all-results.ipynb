```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
city = "mucuri"
Name = "Mucuri"
height = '100'

plt.figure(figsize=(10,7), dpi=320, facecolor='w', edgecolor='k')
plt.title(f"{Name} {height}m Val Loss by depth")
plt.xlabel("Epoch")
plt.ylabel("Val Loss")
plt.grid()
for n_layers in range(1,9):
    history = pd.read_csv(f"analysis/loss-{city}-{height}-{n_layers}-layers.csv")
    plt.plot(history['Val Loss'], label=f"Val Loss depth {n_layers}")
    plt.xticks(range(0, len(history['Loss'])+1, 5))
    plt.ylim(top=50, bottom=0)
    #plt.yticks(range(0, 21, 5))
    plt.legend(loc='best')
    #plt.plot(history['Val Loss'], label="Val Loss/Epoch")
```

```python
plt.figure(figsize=(14,5), dpi=320, facecolor='w', edgecolor='k')
plt.title(f"Loss by depth")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid()
heights = ["100", "120", "150"]
for n_layers in range(1,9):
    history = pd.read_csv(f"analysis/loss-esplanada-120-{n_layers}-layers.csv")
    plt.plot(history['Loss'], label=f"Loss depth {n_layers}")
    plt.xticks(range(0, len(history['Loss'])+1, 5))
    plt.legend(loc='best')
    #plt.plot(history['Val Loss'], label="Val Loss/Epoch")
```

```python
df100 = pd.read_csv("analysis/metrics-esplanada-100.csv", index_col=0)
df120 = pd.read_csv("analysis/metrics-esplanada-120.csv", index_col=0)
df150 = pd.read_csv("analysis/metrics-esplanada-150.csv", index_col=0)

df_concat = pd.concat((df100, df120, df150))
by_row_index = df_concat.groupby(df_concat.index)
df_means = by_row_index.mean()
df_means[["RMSE", "NRMSE", "R", "Fator de 2"]]
```
